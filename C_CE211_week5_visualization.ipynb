# Univariate Visualization:  Visualization for single variable

**Bar Chart** for qualitative (categorical) variables
Categorical variables are basically the variables that are discrete and not continuous. Ex: color of an item is a discrete variable whereas its price is a continuous variable.



*Categorical variables are further divided into 2 types:*

* Ordinal categorical variables These variables can be ordered. Ex: Size of a T shirt (M < L< XL)
* Nominal categorical variables These variables can t be ordered. Ex: Color of a T shirt. We can t say that Blue Green as it doesn t make any sense to

```python
data_size = f_data.shape[0]
#num_airlines= len(f_data['airline'].value_counts().index)
max_count_airline = f_data['airline'].value_counts()[0]

max_prop = max_count_airline / data_size
max_prop   # %of the airline that has max num of flight
```

```python
import numpy as np
prop_list = np.arange(0,max_prop, 0.05)
prop_list
```

```python
sns.countplot(data = f_data, y='airline', color = base_color, order = order_list);
plt.xticks(prop_list*data_size, ['{:.2f}'.format(i) for i in prop_list] )      # the x-axis labels was rotation
plt.xlabel('Percentage')
```

```python
sns.countplot(data = f_data, y='airline', color = base_color, order = order_list);
plt.xticks(prop_list*data_size, ['{:.2f}'.format(i) for i in prop_list] )      # the x-axis labels was rotation
plt.xlabel('Percentage')

for i in range (0,order_list.shape[0]):
    count = f_data['airline'].value_counts()[i]
    text = '{:.1f}%'.format((100*count)/data_size)
    plt.text(count+1, i, text)   #(position_x, position_y ,written_text)
```

# Bivariate Visualization:  Visualization for two variables


```python
t_data = pd.read_csv('titanic.csv')
t_data.head()
```

```python
t_data.isnull().sum()
```

```python
t_data.drop('Cabin', axis=1, inplace=True)
t_data['Age'].fillna(t_data['Age'].mean(),inplace=True)
t_data.dropna(inplace=True)
```

## **Scatter plots** (Quantitative variable VS. Quantitative variable)



We can use *matplotlib.pyplot* function that called **scatter** or *seaborn* function that called **regplot**

```python
plt.figure(figsize=(10,6))
plt.scatter(data = t_data, x ='Age', y='Fare')
plt.xlabel('Age')
plt.ylabel('Fare');
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = t_data, x ='Age', y='Fare', line_kws = {"color": "red"})
plt.xlabel('Age')
plt.ylabel('Fare');

```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = t_data, x ='Age', y='Fare',fit_reg=False)
plt.xlabel('Age')
plt.ylabel('Fare');
```

# Overplotting happens because too many overlapping points

```python
sns.regplot(data = t_data, x ='Pclass', y='Age',fit_reg=False)
plt.xlabel('Pclass')
plt.ylabel('Age');
```

```python
# We can solve it with Jitter
sns.regplot(data = t_data, x ='Pclass', y='Age', x_jitter=0.4,fit_reg=False)
plt.xlabel('Pclass')
plt.ylabel('Age');
```

```python
# Also, we can solve it with Transparency
sns.regplot(data = t_data, x ='Pclass', y='Age', scatter_kws={'alpha':0.2},fit_reg=False)
plt.xlabel('Pclass')
plt.ylabel('Age');
```

## Box Plot



**Example:** The number of books taken out of the library per month is as follows:
            
            3, 0, 12, 0, 2, 0, 26, 0, 7, 5, 5, 2, 1, 1, 2

*Step_1:*   Sort Data   0, 0, 0, 0, 1, 1, 2, 2, 2, 3, 5, 5, 7, 12, 26

*Step_2:*   **min=0    max=26   Q1=0       Q3=5**

*Step_3:*   **IQR = Q3 - Q1** = 5 - 0 = 5

*Step_4:*   **LowerValue = min - (1.5 * IQR)** = 0 - 1.5*5 = -7.5
            **UpperValue = max + (1.5 * IQR)** = 5 + 1.5*5 = 12.5
            
*Step_5:*  Any point less than -7.5 or greater than 12.5 will be **Outlier**



```python
sns.boxplot(data= t_data, x= 'Pclass', y='Age');
```
